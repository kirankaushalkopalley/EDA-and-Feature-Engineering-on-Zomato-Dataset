__df.isnull().sum()__ is the function used to find the number of missing values in each and every column of our dataset.

```python
df.isnull().sum()
```

Another way of identifying the columns which have atleast one missing value in it.

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

Identifying the columns which have missing values using a visualization.

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

"Country Code" is the common column in both the dataframes and can be used to <u>JOIN</u> these two datasets.

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(5)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

To understand how many transactions have been done in each country sorted from highest to lowest.

```python
final_df.Country.value_counts()
```

From the above output, if we need to plot a Pie Chart for top 3 countries, then the country names have to saved to a variable separately and values need to be stored separately as shown below:

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

The __"autopct"__ property within the pie function is used to show the percentage value instead of the actual value. __%1.2f%%__ denotes that after the decimal point we want to see two values.

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

### Observation-1:

* Zomato maximum records or transaction are from India 
* After that USA and then United Kingdoms

```python
final_df.columns
```

The __Aggregate rating__ feature looks to be interesting. Lets analyze that..!!

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation-2

* When the Rating is between 4.5 to 4.9  --> __Excellent__
* When the Rating is between 4.0 to 4.4 --> __Very Good__
* When the Rating is between 3.5 to 3.9 --> __Good__
* When the Rating is between 2.5 to 3.4 --> __Average__
* When the Rating is between 2.0 to 2.4 --> __Poor__
* __None of the customers rated 5.0__

```python
ratings.head()
```

A <u>Bar Chart</u> for __Rating Count Vs Aggregate rating__

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

__To make the visualizations look bigger__

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

Using the given "Rating Colors" for the "Rating Counts" for improving the readability.

```python
sns.set_theme(style="whitegrid", rc={"axes.facecolor": "white"})
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['black','red','orange','yellow','green','green'])
```

### Observation-3:
* People not rating the App is more.
* Maximum number of rating are between 2.5 to 3.4.

The frequency of ratings under each Rating Color is shown using __Count Plot__.

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
